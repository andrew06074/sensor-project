```python
import pandas as pd
import numpy as np

df = pd.read_csv('sensor-sl.csv')
```

```python
df
```

```python
for item in df['Time']:
    time = item
    hours = int(time)
    minutes = (time*60) % 60
    seconds = (time*3600) % 60
    new_time = ("%d:%02d.%02d" % (hours, minutes, seconds))
    df['Time'] = df['Time'].replace(new_time)
```
